```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

### **Scatterplot - Numerical-Numerical combination**

```python
# Scatterplot of total bill vs tip paid

sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
plt.show()
```

**Seperate the scattered points based on categorical data (hue)**

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
plt.show()
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
plt.show()
```

### **Barplot - Numerical-Categorical combination**

```python
sns.barplot(x=titanic['pclass'],y=titanic['fare'])
plt.show()
```

```python
sns.barplot(x=titanic['pclass'],y=titanic['fare'],hue=titanic['sex'])
plt.show()
```

### **Box plot - Numerical-Categorical combination**

```python
sns.boxplot(x=titanic['sex'],y=titanic['age'])
plt.show()
```

```python
sns.boxplot(x=titanic['sex'],y=titanic['age'],hue=titanic['survived'])
plt.show()
```

### **Distplot - Numerical categorical combination**

```python
# Distribution of ages of people who survived
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)

# Distribution of ages of people who did not survive
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)

plt.show()
```

### **Heatmap - Categorical-Categorical combination**

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
plt.show()
```

```python
titanic.groupby('pclass').mean()['survived']
```

```python
titanic.groupby('embarked').mean()['survived']
```

### **Clustermap - Categorical-Categorical combination**

```python
pd.crosstab(titanic['sibsp'],titanic['survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['sibsp'],titanic['survived']))
plt.show()
```

### **Pairplot**

```python
sns.pairplot(iris,hue='species')
plt.show()
```

### **Lineplot - Numerical-Numerical combination (for time based)**

```python
df_flights = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(x=df_flights['year'],y=df_flights['passengers'])
```
